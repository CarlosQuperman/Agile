### 4. 정보의 시각화

- 이제 우리는 df_new안에 들어있는 정보를 토대로
- 경기도내 특수학교 지도를 만들어보자
- 지도를 만들기 위해서는 어떻게 해야 할까?
- df_new를 처음부터 끝까지 한줄씩 돌면서 위도와 경도값을 이용하여 지도에 표시해주고 (위도와 경도가 0이 아니면)
- 지도를 표시해주는 마커에 정보를 담아주면 된다.

(1) 지도 정보 확인하기

- 지도 시각화의 구조를 알기 위해
- 수원 화성 팔달문의 위도, 경도와 정보를 입력하면
- 이를 지도에 표시하는 시각화를 진행해보자.

```python
import folium # 포리움 라이브러리

m = folium.Map([37.278230094578,127.02271824041]) #지도의 시작값을 지정하고 만들어줌
folium.Marker([37.278230094578,127.02271824041], popup='수원화성', tooltip="팔달문").add_to(m) ### 팝업은? (마우스 올렸을 때) 툴팁은? (마우스 클릭했을 때)
m
```

(2) 한줄씩 돌면서 위도와 경도 값에 따라 마커를 만들어주고 누르면 주소가 마우스 올려놓으면 가게 이름이 뜨는 지도 만들기

- 이제 df_new를 한줄씩 돌면서
- 위도와 경도값을 가져와 저장하고 (위도와 경도가 0이 아니면)
-마우스를 올려두면 학교명이 나타난다
-클릭하면 도로명주소가 나타난다.

```python
m = folium.Map([37.278230094578,127.02271824041],zoom_start=20) #지도의 시작값과 확대 정도를 지정하고 만들어줌

for i in df_new.index : #위치번호(index)로 처음부터 끝까지 돌면서
  if df_new.loc[i,'WGS84위도']!='0' and df.loc[i,'WGS84경도']!='0': ### 위도와 경도가 둘 다 0이 아니면을 나타내도록
    ### 우리가 필요한 정보는 위도, 경도, 사업장명, 소재지도로명주소
    lat = df_new.loc[i,'WGS84위도']
    lon = df_new.loc[i,'WGS84경도']
    title = df_new.loc[i,'학교명']
    address = df_new.loc[i,'소재지도로명주소']

    icon_color = 'blue' ###색지정(파란색으로)

    folium.Marker([lat,lon], popup=address, tooltip=title,icon=folium.Icon(color=icon_color)).add_to(m)

m
```

(3) 공립, 사립, 국립의 색깔 구별해주기

- 우리가 위에서 만든 지도는 학교의 색깔이 똑같아서 구별하기가 쉽지 않다.
- 사립은 파란색으로, 국립은 초록색으로, 공립은 빨간색으로  색깔을 다르게 설정해주는 프로그래밍을 수행해보자.

```python
m = folium.Map([37.278230094578,127.02271824041],zoom_start=20) #지도의 시작값과 확대 정도를 지정하고 만들어줌

for i in df_new.index : #위치번호(index)로 처음부터 끝까지 돌면서
  if df_new.loc[i,'WGS84위도']!='0' and df.loc[i,'WGS84경도']!='0':
    ### 우리가 필요한 정보는 위도, 경도, 상호명, 주소
    lat = df_new.loc[i,'WGS84위도']
    lon = df_new.loc[i,'WGS84경도']
    title = df_new.loc[i,'학교명']
    brand = df_new.loc[i,'학교유형명'] #통일된 브랜드명이 필요해서 설정
    address = df_new.loc[i,'소재지도로명주소']

    icon_color = 'blue' ###색지정(파란색으로)

    ### 만약 브랜드명이 뚜레쥬르면 색깔을 녹색(green)으로 지정
    if brand == '국립':
      icon_color = 'green'

    if brand == '공립':
        icon_color = 'red'

    folium.Marker([lat,lon], popup=address, tooltip=title,icon=folium.Icon(color=icon_color)).add_to(m)

m
```
